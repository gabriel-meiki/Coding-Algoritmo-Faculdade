```python
# são 5 patinhos, a cada vez vai indo embora 1 pato, quando todos os patos vão embora, a mãe patinho encontra todos os patos.

patinhos = 5

while patinhos > 0:
  print(f'{patinhos} patinhos foram passear, além das montanhas para brincar. A mamãe gritou Quá Quá Quá.')
  patinhos -= 1
  print(f'Mas só {patinhos} patinhos voltaram de lá.')
  print('_' * 10)
```

```python
patinhos = [5, 4, 3, 2, 1]

for patinho in patinhos:
  print(f'{patinho} patinhos foram passear, além das montanhas para brincar. A mamãe gritou Quá Quá Quá.')
  print(f'Mas só {patinho - 1} patinhos voltaram de lá.')
  print('_' * 10)
```
